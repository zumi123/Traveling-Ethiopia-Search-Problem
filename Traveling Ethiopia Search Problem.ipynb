1.2 Write a class that takes the converted state space graph, initial state, goal state and a search strategy and return       the corresponding solution/path according to breadth-first search and depth-first search strategies.

```python
from collections import deque

class StateSpaceGraph:
    def __init__(self, graph):
        self.graph = graph

    def bfs(self, initial_state, goal_state):
        queue = deque([(initial_state, [initial_state])])
        visited = set()

        while queue:
            current_state, path = queue.popleft()
            if current_state == goal_state:
                return path
            visited.add(current_state)

            for neighbor in self.graph.get(current_state, []):
                if neighbor not in visited:
                    queue.append((neighbor, path + [neighbor]))

        return None  # Goal state not reachable

    def dfs(self, initial_state, goal_state):
        stack = [(initial_state, [initial_state])]
        visited = set()

        while stack:
            current_state, path = stack.pop()
            if current_state == goal_state:
                return path
            visited.add(current_state)

            for neighbor in reversed(self.graph.get(current_state, [])):
                if neighbor not in visited:
                    stack.append((neighbor, path + [neighbor]))

        return None  # Goal state not reachable

    
```

```python
# Create an instance of the StateSpaceGraph class
ethiopia = StateSpaceGraph(ethiopia_graph)

# Define initial and goal states
initial = 'Addis Ababa'
goal = 'Lalibela'

# Perform BFS
bfs_path = ethiopia.bfs(initial, goal)
print(f"BFS Path from {initial} to {goal}: {bfs_path}")

# Perform DFS
dfs_path = ethiopia.dfs(initial, goal)
print(f"DFS Path from {initial} to {goal}: {dfs_path}")
```

2.2. Assuming “Addis Ababa” as an initial state, write a program that use uniform cost search
to generate a path to “Lalibela”.

```python
import heapq

def uniform_cost_search(graph, start, goal):
    # Priority queue, stores tuples with cumulative cost and current path
    priority_queue = [(0, [start])]
    # Set to keep track of visited nodes
    visited = set()

    while priority_queue:
        # Pop the path with the smallest cost
        cost, path = heapq.heappop(priority_queue)
        current = path[-1]
        # Check if the current node is the goal
        if current == goal:
            return (cost, path)
        if current not in visited:
            visited.add(current)
            # Explore neighbors
            for neighbor, edge_cost in graph[current]:
                if neighbor not in visited:
                    # Add the new path with updated cost and path
                    heapq.heappush(priority_queue, (cost + edge_cost, path + [neighbor]))
    return None

# Call the search algorithm
cost, path = uniform_cost_search(ethiopia_graph_with_costs, 'Addis Ababa', 'Lalibela')
print(f"Path found: {path} with total cost: {cost}")
```

2.3. Given “Addis Ababa” as an initial state and “Axum”, “Gondar”, “Lalibela”, "Babile", “Jimma”, “Bale”, “Sof Oumer”, and “Arba Minch” as goal states;in no specific order, write a customized uniform cost search algorithm to generate a path that let a visitor visit all those goal states preserving the local optimum.

```python
import heapq

def customized_uniform_cost_search(graph, start, goals):
    # Priority queue, stores tuples with cumulative cost, current path, and remaining goals
    priority_queue = [(0, [start], set(goals))]
    
    # Keep track of the visited states and the best cost to reach them
    visited = set()
    
    while priority_queue:
        # Pop the path with the smallest cost
        cost, path, remaining_goals = heapq.heappop(priority_queue)
        current = path[-1]
        
        # If we reach a goal that is in our remaining goals, we remove it from the remaining_goals set
        if current in remaining_goals:
            remaining_goals.remove(current)
        
        # If we have visited all goals, return the current path and cost
        if not remaining_goals:
            return cost, path
        
        # Prevent cycles by checking if we have visited this node with this set of remaining goals
        visited_state = (current, tuple(sorted(remaining_goals)))
        if visited_state not in visited:
            visited.add(visited_state)
            
            # Explore neighbors
            for neighbor, edge_cost in graph.get(current, []):
                if neighbor not in path:  # Prevent loops in the path
                    # Add the new path with updated cost and remaining goals
                    heapq.heappush(priority_queue, (cost + edge_cost, path + [neighbor], remaining_goals.copy()))
    return None

# Example usage
start_city = 'Addis Ababa'
goal_cities = ['Axum', 'Gondar', 'Lalibela', 'Babile', 'Jimma', 'Bale', 'Sof Oumer', 'Arba Minch']
cost, path = customized_uniform_cost_search(ethiopia_graph_with_costs, start_city, goal_cities)
print(f"Path found: {path} with total cost: {cost}")
```

```python
import heapq

class AStarSearch:
    def __init__(self, start, goal, graph):
        self.start = start
        self.goal = goal
        self.graph = graph
        self.open_set = []   
        heapq.heappush(self.open_set, (self.heuristic(start), start))  # Priority queue
        self.came_from = {}  # Tracks the best path
        self.g_score = {node: float('inf') for node in graph}
        self.g_score[start] = 0
        self.f_score = {node: float('inf') for node in graph}
        self.f_score[start] = self.heuristic(start)

    def heuristic(self, node):
        # This function should return the heuristic estimate from `node` to the goal.
        return self.graph[node]['heuristic']

    def reconstruct_path(self, current):
        path = [current]
        while current in self.came_from:
            current = self.came_from[current]
            path.insert(0, current)
        return path

    def search(self):
        while self.open_set:
            current_f_score, current = heapq.heappop(self.open_set)
            if current == self.goal:
                return self.reconstruct_path(current)

            for neighbor, backward_cost in self.graph[current]['neighbors']:
                tentative_g_score = self.g_score[current] + backward_cost
                if tentative_g_score < self.g_score[neighbor]:
                    self.came_from[neighbor] = current
                    self.g_score[neighbor] = tentative_g_score
                    self.f_score[neighbor] = tentative_g_score + self.heuristic(neighbor)
                    if neighbor not in [item[1] for item in self.open_set]:
                        heapq.heappush(self.open_set, (self.f_score[neighbor], neighbor))
        return None

astar = AStarSearch(start='Addis Ababa', goal='Moyale', graph=graph)
path = astar.search()
print(f"Path found: {path}")
```

```python
class MiniMaxSearch:
    def __init__(self, ethiopia_graph):
        self.ethiopia_graph = ethiopia_graph  # A dictionary representing the ethiopia_graph

    def minimax(self, node, depth, is_maximizing_player, is_adversary=False):
        if depth == 0 or self.is_terminal(node):
            utility = self.get_utility(node)
            if utility is None:
                return float('-inf') if is_maximizing_player else float('inf')
            return utility

        if is_maximizing_player:
            best_value = float('-inf')
            for neighbor in self.get_neighbors(node):
                value = self.minimax(neighbor, depth - 1, False, not is_adversary)
                best_value = max(best_value, value)
            return best_value
        else:
            best_value = float('inf')
            for neighbor in self.get_neighbors(node):
                value = self.minimax(neighbor, depth - 1, True, not is_adversary)
                best_value = min(best_value, value)
            return best_value


    def get_utility(self, node):
        # Return the utility value of a terminal node
        return self.ethiopia_graph[node].get('utility')

    def get_neighbors(self, node):
        # Return the neighbors of a node
        return self.ethiopia_graph[node]['neighbors']
    
    def get_adversary(self, node):
        # Return the adversary value of a node
        return self.ethiopia_graph[node]['adversary']

    def is_terminal(self, node):
        # Check if the node is a terminal node
        return 'utility' in self.ethiopia_graph[node]

    def find_best_move(self, start_node):
        best_utility = float('-inf')
        best_move = None

        for neighbor in self.get_neighbors(start_node):
            utility = self.minimax(neighbor, depth=4, is_maximizing_player=False)
            is_adversary= self.get_adversary(neighbor)
            if utility is not None and utility > best_utility and is_adversary is not True:
                best_utility = utility
                best_move = neighbor

        if best_move is None:
            # Handle the case where no best move is found
            print("No best move found within the specified depth.")

        return best_move


minimax_search = MiniMaxSearch(ethiopia_graph)
best_move = minimax_search.find_best_move('Addis Ababa')
print(f"The best move for the agent is to go to: {best_move}")
```
